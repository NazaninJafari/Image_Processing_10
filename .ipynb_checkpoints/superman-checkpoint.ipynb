```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('imgs/SuperMan.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
img_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

H, S, V = cv2.split(img_hsv)
```

```python
bg_img = cv2.imread('imgs/gratte-ciel.jpg')
bg_img = cv2.cvtColor(bg_img, cv2.COLOR_BGR2RGB)
plt.imshow(bg_img)
```

```python
rows, cols = H.shape

bg_img = cv2.resize(bg_img, (cols,rows))
```

```python
for i in range(rows):
    for j in range(cols):
        if 35<H[i,j]<75 and S[i,j]>50 and V[i,j]>80:
            image[i,j] = bg_img[i,j]
```

```python
plt.imshow(image)
```

```python
result = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
cv2.imwrite('imgs_result/superman_result.jpg' ,result)
```
